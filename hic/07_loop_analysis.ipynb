```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import LogNorm
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import bioframe
```

```python
def get_loop_strength(pile):
    loop_strength = np.mean(pile[19:22, 19:22])
    return(str(round(loop_strength, 2)))
```

```python
pile1 = pd.read_csv('./loops/average_loops_T-MAU2.txt', skiprows = 35, sep = '\s+', header = None).values
pile2 = pd.read_csv('./loops/average_loops_T-MAU2_Dox_1H.txt', skiprows = 35, sep = '\s+', header = None).values
pile3 = pd.read_csv('./loops/average_loops_T-mCherry.txt', skiprows = 35, sep = '\s+', header = None).values
pile4 = pd.read_csv('./loops/average_loops_haarhuis_2017.WT.txt', skiprows = 35, sep = '\s+', header = None).values
pile5 = pd.read_csv('./loops/average_loops_sanborn_2015.WT.txt', skiprows = 35, sep = '\s+', header = None).values
```

```python
plt.figure(figsize = (10, 4))
plt.suptitle('Loops (N=8096)', fontsize = 16)

plt.subplot(131)
plt.title('T-MAU2', fontsize = 14)
plt.imshow(np.log2(pile1), vmin = -1.2, vmax = 1.2, cmap = 'coolwarm')
plt.text(x = 39, y = 2, s = get_loop_strength(pile1), ha = 'right', va = 'top', fontsize = 10)
cbar = plt.colorbar(shrink = 0.5)
cbar.set_label('log$_{2}$(contact frequency obs/exp)', fontsize = 8)
plt.xticks([], [])
plt.yticks([], [])

plt.subplot(132)
plt.title('T-MAU2 + Dox 1H', fontsize = 14)
plt.imshow(np.log2(pile1), vmin = -1.2, vmax = 1.2, cmap = 'coolwarm')
plt.text(x = 39, y = 2, s = get_loop_strength(pile2), ha = 'right', va = 'top', fontsize = 10)
cbar = plt.colorbar(shrink = 0.5)
cbar.set_label('log$_{2}$(contact frequency obs/exp)', fontsize = 8)
plt.xticks([], [])
plt.yticks([], [])

plt.subplot(133)
plt.title('T-mCherry', fontsize = 14)
plt.imshow(np.log2(pile3), vmin = -1.2, vmax = 1.2, cmap = 'coolwarm')
plt.text(x = 39, y = 2, s = get_loop_strength(pile3), ha = 'right', va = 'top', fontsize = 10)
cbar = plt.colorbar(shrink = 0.5)
cbar.set_label('log$_{2}$(contact frequency obs/exp)', fontsize = 8)
plt.xticks([], [])
plt.yticks([], [])

plt.tight_layout()
plt.savefig('/DATA/users/magnitov/tacl/paper_figures_new/loops_1.pdf', dpi = 300, bbox_inches = 'tight')
plt.show()
```
